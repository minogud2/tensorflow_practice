### Cancer Classification

```python
# basic imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset
df = pd.read_csv('cancer_classification.csv')
```

```python
# describe dataset to check for datatypes, null values and columns.
df.info()
```

```python
# Obtain descriptive stats on the dataset. 
df.describe().transpose()
```

```python
sns.countplot(x='benign_0__mal_1', data=df)

# Overall there are more malignant classifications than benign but overall, it's relatively unbiased. 
```

```python
plt.figure(figsize=(12,6))
df.corr()['benign_0__mal_1'][:-1].sort_values().plot(kind='bar')

# Features are almost all negatively correlated with the target variable. High coorrelations means the model is likely to 
# produce more accurate predictions. 
```

```python
# plot out the correlations
plt.figure(figsize=(12,12))
sns.heatmap(df.corr())
```

```python
# split the dataset to create the data model. 
# Tensorflow doesn't allow pandas dataframes as inputs and therefore values need to be extracted. 
X = df.drop('benign_0__mal_1', axis=1).values
y= df['benign_0__mal_1'].values
```

```python
# Split the dataset into a training and test split.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
df.shape
```

```python
# Tensorflow only allows scaled data as input. Therefore importing min-max scaling to transform the data. 
from sklearn.preprocessing import MinMaxScaler
```

```python
# Only fit to the training data in order to prevent data leakage. Otherwise the model would overfit. 
scaler=MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

# Starting point for number of nodes should be number of features.
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

# Output is classification- so sigmoid function required to predict between 0 and 1. 
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses_df = pd.DataFrame(model.history.history)
```

```python
# From the plot below you can see there are too many epochs. The validation loss is getting higher from approx 35 onwawrds. 
losses_df.plot()
```

```python
# Will use early stopping to identify the best stopping point for the model to minimise errors. 
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

# Output is classification- so sigmoid function needed.
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
# Since we are looking to monitor the validation loss, the mode is to minimize the error. If it was more important to 
# maximise accuracy, that would be monitored adn the mode =max. 
# loss of zero is a perfect fit.
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop])

# Choosing an arbitraty large number, with callbacks finding the optimal epochs
```

```python
# stopped at 89 instead of 600. 

model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
# can further look to optimize by using dropout which will drop/turn off neurons randomly in an attempt to increase accuracy
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
# Dropout should be between 0-1. e.g. 0.5 means half of neurons will be turned off randomly. 
model.add(Dense(15, activation='relu'))

# Output is classification- so sigmoid function needed.
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop])
```

```python
# Model is performing better with loss and val_loss converging between 60-80. 
model_loss2 = pd.DataFrame(model.history.history)
model_loss2.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
# check accuracy of data model.
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
# Model performs very well with a high accuracy. 
# Accuracy is a good metric because the target classes trained were relatively equal.
```
